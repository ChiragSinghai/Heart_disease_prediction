## data preprocessing

```python
X,Y=df.loc[:,:'thal'],df['target']
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
print ("train_set_x shape: " + str(X_train.shape))
print ("train_set_y shape: " + str(Y_train.shape))
print ("test_set_x shape: " + str(X_test.shape))
print ("test_set_y shape: " + str(Y_test.shape))
```

```python
X_train0,X_test0,Y_train0,Y_test0=train_test_split(X,Y,test_size=0.3)
print ("train_set_x shape: " + str(X_train0.shape))
print ("train_set_y shape: " + str(Y_train0.shape))
print ("test_set_x shape: " + str(X_test0.shape))
print ("test_set_y shape: " + str(Y_test0.shape))
```

# Decision Tree

```python
from sklearn.metrics import confusion_matrix, classification_report 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
```

```python
prediction=dt.predict(X_test)
accuracy_dt=accuracy_score(Y_test,prediction)*100
accuracy_dt
#dt.score(X_test,Y_test)
#print(X_test.shape,Y_test.shape)
```

```python
print("Feature importances:\n{}".format(dt.feature_importances_))
```

```python
dt_1=DecisionTreeClassifier()
dt_1.fit(X_train0,Y_train0)
predic=dt_1.predict(X_test0)
accuracy=accuracy_score(Y_test0,predic)*100
accuracy
```

```python
def plot_feature_importances_diabetes(model,data):
    n_features=data.shape[1]
    
    plt.figure(figsize=(8,6))
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features),data)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
    plt.ylim(-1, n_features)
plot_feature_importances_diabetes(dt,X_train)
plt.savefig('feature_importance')
```

```python
X_train1=X_train.drop('fbs',axis=1)
X_test1=X_test.drop('fbs',axis=1)
print ("train_set_x shape: " + str(X_train1.shape))
print ("train_set_y shape: " + str(Y_train.shape))
print ("test_set_x shape: " + str(X_test1.shape))
print ("test_set_y shape: " + str(Y_test.shape))
dt1=DecisionTreeClassifier()
dt1.fit(X_train1,Y_train)
```

```python
pre=dt1.predict(X_test1)
a=accuracy_score(pre,Y_test)*100
a
```

```python
cm = confusion_matrix(Y_test,pre) 
conf_matrix = pd.DataFrame(data = cm,  
                           columns = ['Predicted:0', 'Predicted:1'],  
                           index =['Actual:0', 'Actual:1']) 
plt.figure(figsize = (8, 5)) 
sns.heatmap(conf_matrix, annot = True, fmt = 'd', cmap = "Greens") 
plt.show() 
#plt.savefig('Accuracy_decision')
```

```python
plot_feature_importances_diabetes(dt1,X_train1)
```
